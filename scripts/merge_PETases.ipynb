Script para hacer merge entre la tabla def de EXTREMÓFILOS con la nueva tabla con todas las PETases

```python
import pandas as pd
import numpy as np
```

```python
PETases = pd.read_excel("../tablas_PETases/total_PETases.xlsx", sheet_name="total_PETases")
PETases.head()
```

```python
tabla = pd.read_excel("../EXTREMOPHILES_REFINED_31MAR2026_CompletnessContamination.xlsx", sheet_name="Genomes")
tabla.head()
```

Quiero copiar la información de los genomas de la tabla de extremófilos pero solo los genomas que están presentes en mi tabla de PETases.
Primero como prueba copio solo la localización

```python
PETases = PETases.merge(tabla[["NCBI acc", "Location.1"]], on="NCBI acc", how="left")
PETases.head()
```

```python
# Muevo los datos de la columna creada a mi columna Location
PETases["Location"] = PETases["Location.1"]
# Quito la columna extra
PETases = PETases.drop(columns=["Location.1"])
PETases.head()
```

Hacemos un merge con el resto de las columnas que faltan. 
Como me va a añadir las columnas de nuevo al estar repetidas voy a quitarlas primero.

```python
PETases = PETases.drop(columns=[
    "Latitude", "Longitude", "Isolation Source",
    "true_extreme_environment", "number_extr_environments",
    "type_env_extr", "environment", "env comments",
    "Final especie",
    "true_micr_extreme",
    "microorg_extremophile",
    "microorg cooments"
], errors="ignore")
```

```python
columnas = ["NCBI acc", "Latitude", "Longitude", "Isolation Source",
    "true_extreme_environment", "number_extr_environments",
    "type_env_extr", "environment", "env comments",
    "Final especie", "true_micr_extreme", "microorg_extremophile", "microorg cooments"]

PETases = PETases.merge(tabla[columnas], on="NCBI acc", how="left")
PETases.head(20)
```

```python
# Guardamos el excel
PETases.to_excel("PETases_03ABR26.xlsx", index=False)
```
